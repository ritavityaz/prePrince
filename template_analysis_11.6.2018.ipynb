## Compare the results generated by find_templates.py and the ones that are currently used for Prince.

H37Rv was used as a reference genome.

Note: templates 802 and 2687 only have one copy in the reference.

```python
from Bio import SeqIO

found_templates = list(SeqIO.parse("found_templates.fasta", "fasta"))
original_templates = list(SeqIO.parse("templates.fasta", "fasta"))

record_dict = SeqIO.to_dict(SeqIO.parse("found_templates.fasta", "fasta"))

# print(found_templates)
# print(record_dict["580"].seq)


for i, template in enumerate(original_templates):
    name = template.id
    identical = template.seq == record_dict[name].seq
    print(str(name) +"          "+ str(identical))

    print(record_dict[name].seq)
    print(template.seq)
    print(" ")
  
```

Observation 1: Some of the templates that did not fully match are only different in a couple of nucleotides.

Observation 2: For template 4052, top template has a lot consecutive A's, and the bottom a lot of consecutive T's. Could one be the reverse complement of the other?

Goal 2: Align them to evaluate how different they really are.

```python
from Bio import pairwise2

alignments = pairwise2.align.localds

for i, template in enumerate(original_templates):
    name = template.id
    identical = template.seq == record_dict[name].seq
    print(str(name) +"          "+ str(identical))

    alignment = pairwise2.align.localxx(template.seq,record_dict[name].seq)
#     print(alignment)
    print(pairwise2.format_alignment(*alignment[0]))
    
    
    
#     print(str(name) +"          "+ str(identical))

#     print(record_dict[name].seq)
#     print(template.seq)
#     print(" ")
  

```
